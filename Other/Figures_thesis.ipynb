**Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Plot Activation Functions**

```python
# Input range
x = np.linspace(-6, 6, 1000)
```

```python
# Sigmoid function
y = 1 / (1 + np.exp(-x))

# Plot Sigmoid
plt.figure()
plt.plot(x, y, color='deepskyblue')
plt.title('Sigmoid')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.5)
plt.show()
```

```python
# ReLU function
y = np.maximum(0, x)

# Plot ReLU
plt.figure()
plt.plot(x, y, color='deepskyblue')
plt.title('ReLU')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.5)
plt.show()
```

```python
# GELU function
y = 0.5 * x * (1 + np.tanh(np.sqrt(2/np.pi) * (x + 0.044715 * x**3)))

# Plot GELU
plt.figure()
plt.plot(x, y, color='deepskyblue')
plt.title('GELU')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.5)
plt.show()
```
